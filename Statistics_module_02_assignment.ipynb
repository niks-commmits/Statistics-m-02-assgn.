1.  What is hypothesis testing in statistics?
  -  Hypothesis testing is a statistical method used to make
     decisions or inferences about a population based on sample data.

2.  What is the null hypothesis, and how does it differ from  
   the alternative hypothesis?

  -  The null hypothesis (HO)and the alternative hypothesis (HA)
     are fundamental concepts in hypothesis testing. They define the competing claims we evaluate using statistical methods.

3. What is the significance level in hypothesis testing, and  
   why is it important?

   -  The significance level (α) is the probability of  
      rejecting the null hypothesis (HO) when it is actually true.
      The significance level helps determine the threshold (critical value) for rejecting H0 If the p-value (calculated from the test) is less than or equal to α, we reject H 0.

4.  What does a P-value represent in hypothesis testing?

  -  A p-value is the probability of obtaining a test result  
     at least as extreme as the one observed, assuming that the null hypothesis (HO) is true. It helps determine whether the observed data provide enough evidence to reject HO .

5. How do you interpret the P-value in hypothesis testing?

  -  The p-value tells us how likely we are to get the observed
     results (or more extreme ones) if the null hypothesis (HO) is true. It helps us decide whether to reject or fail to reject HO.
6. What are Type 1 and Type 2 errors in hypothesis testing?

  -  In hypothesis testing, we make decisions based on sample
     data, but there's always a chance of making an incorrect conclusion. These incorrect conclusions are called Type I and Type II errors.

7. What is the difference between a one-tailed and a two-tailed
   test in hypothesis testing?

   - When conducting hypothesis tests, we need to decide  
     whether we are testing for a difference in one direction or in either direction. This determines whether we use a one-tailed or two-tailed test.

8. What is the Z-test, and when is it used in hypothesis  
   testing?
   - A Z-test is a statistical test used to determine whether
     the mean of a sample differs significantly from a known population mean. It is based on the Z-score, which measures how many standard deviations a data point is from the mean.
9.  How do you calculate the Z-score, and what does
     it represent in hypothesis testing?

  -  A Z-score (also called a standard score) tells us how  
     many standard deviations a data point or sample statistic is from the population mean. It helps standardize different datasets, making it easier to compare them.

10.  What is the T-distribution, and when should it be used
     instead of the normal distribution?

    - The T-distribution is a probability distribution that is
      similar to the normal distribution but has heavier tails. This means it has a higher probability of producing values that are far from the mean.

11.  What is the difference between a Z-test and a T-test?

    - The Z-test and T-test are both statistical tests used to
      compare means, but they differ in when and how they should be used.
      Use a Z-test when:
      The sample size is large (n>30).
      The population standard deviation (σ) is known.
      The data follows a normal distribution (or is approximately normal for large samples).
      Use a T-test when:
      The sample size is small (n<30).
      The population standard deviation is unknown, so it must be estimated from the sample.
      The data is approximately normally distributed.

12.  What is the T-test, and how is it used in hypothesis
     testing?
     - A T-test is a statistical test used to determine whether
       there is a significant difference between the means of two groups. It is particularly useful when the sample size is small (n<30) and the population standard deviation (σ) is unknown. The test relies on the T-distribution, which accounts for variability in small samples.

13.   What is the relationship between Z-test and T-test in
      hypothesis testing?
     -The Z-test and T-test are both statistical methods used in hypothesis testing to compare means and determine if differences between sample data and population parameters are statistically significant. While they serve similar purposes, they differ in their assumptions and applicability.

14. What is a confidence interval, and how is it used to
    interpret statistical results?

    - A confidence interval (CI) is a range of values, derived
      from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence. It provides an estimate of the uncertainty associated with a sample statistic.
15. What is the margin of error, and how does it affect
    the confidence interval?

    - The margin of error (MOE) is the amount by which the estimated value from a sample may differ from the true population value. It represents the uncertainty or variability in the estimation process and determines the width of a confidence interval (CI).

16.  How is Bayes' Theorem used in statistics, and what is its
     significance?

     - Bayes' Theorem is a fundamental concept in probability and statistics that describes how to update our beliefs (probabilities) based on new evidence. It is widely used in statistical inference, machine learning, medical diagnosis, and decision-making.

17.  What is the Chi-square distribution, and when is it used?

    - The Chi-square (χ²) distribution is a continuous          probability distribution that arises when summing the squares of k independent standard normal random variables. It is widely used in statistical hypothesis testing, particularly in tests involving categorical data.

```python
# 1.  Write a Python program to perform a Z-test for comparing a sample mean to a
#     known population mean and interpret the results.

import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided'):

    sample_mean = np.mean(sample)
    sample_size = len(sample)

    standard_error = population_std / np.sqrt(sample_size)

    z_score = (sample_mean - population_mean) / standard_error

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_score)
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_score)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return z_score, p_value, interpretation

sample_data = [50, 52, 48, 47, 51, 49, 53, 46, 50, 52]
population_mean = 50
population_std = 5
alpha = 0.05
alternative = 'two-sided'

z_score, p_value, interpretation = z_test(sample_data, population_mean, population_std, alpha, alternative)

print(f"Z-score: {z_score:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

```python
# 3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)

population_mean = 50
population_std = 10
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score}")
print(f"Critical Z-value (for alpha={alpha}): ±{critical_value}")

if abs(z_score) > critical_value:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)

population_mean = 50
population_std = 10
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.fill_between(x, y, where=(x <= -critical_value), color='red', alpha=0.5, label='Reject H0 (left tail)')
plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.5, label='Reject H0 (right tail)')
plt.axvline(x=-critical_value, color='black', linestyle='dashed', label=f'Critical Z-value: ±{critical_value:.2f}')
plt.axvline(x=critical_value, color='black', linestyle='dashed')
plt.axvline(x=z_score, color='green', linestyle='solid', label=f'Sample Z-score: {z_score:.2f}')
plt.title('Two-Tailed Z-Test - Decision Region Visualization', fontsize=14)
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend(loc='upper left')

plt.grid(True)
plt.show()

print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score}")
print(f"Critical Z-value (for alpha={alpha}): ±{critical_value}")

if abs(z_score) > critical_value:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
# 5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_errors(pop_mean, pop_std, sample_size, alpha=0.05, true_effect_size=5):


    critical_value = stats.norm.ppf(1 - alpha / 2)


    std_error = pop_std / np.sqrt(sample_size)
    x_values = np.linspace(pop_mean - 4 * std_error, pop_mean + 4 * std_error, 1000)
    y_values = stats.norm.pdf(x_values, loc=pop_mean, scale=std_error)


    rejection_region_left = pop_mean - critical_value * std_error
    rejection_region_right = pop_mean + critical_value * std_error


    alternative_mean = pop_mean + true_effect_size
    y_values_alternative = stats.norm.pdf(x_values, loc=alternative_mean, scale=std_error)


    plt.figure(figsize=(10, 6))


    plt.plot(x_values, y_values, label='Null Hypothesis (H0)', color='blue')


    plt.plot(x_values, y_values_alternative, label='Alternative Hypothesis (Ha)', color='orange')


    plt.fill_between(x_values, y_values, where=(x_values <= rejection_region_left), color='red', alpha=0.5, label='Type 1 Error (False Positive - Reject H0)')
    plt.fill_between(x_values, y_values, where=(x_values >= rejection_region_right), color='red', alpha=0.5)

    plt.fill_between(x_values, y_values_alternative, where=(x_values >= rejection_region_left) & (x_values <= rejection_region_right), color='green', alpha=0.5, label='Type 2 Error (False Negative - Accept H0)')

    plt.axvline(rejection_region_left, color='black', linestyle='dashed', label=f'Critical Z-value: ±{critical_value:.2f}')
    plt.axvline(rejection_region_right, color='black', linestyle='dashed')


    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing', fontsize=14)
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()

    type_1_error = 2 * (1 - stats.norm.cdf(critical_value))

    type_2_error = stats.norm.cdf(rejection_region_left, loc=alternative_mean, scale=std_error) + \
                   (1 - stats.norm.cdf(rejection_region_right, loc=alternative_mean, scale=std_error))

    print(f"Type 1 Error (α): {type_1_error:.4f}")
    print(f"Type 2 Error (β): {type_2_error:.4f}")
    print(f"Power of the test (1 - β): {1 - type_2_error:.4f}")

pop_mean = 50
pop_std = 10
sample_size = 30
alpha = 0.05
true_effect_size = 5
visualize_errors(pop_mean, pop_std, sample_size, alpha, true_effect_size)
```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy import stats

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")
```

```python
# 7.  Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
pre_test = np.random.normal(loc=75, scale=10, size=30)
post_test = pre_test + np.random.normal(loc=5, scale=5, size=30)

t_stat, p_value = stats.ttest_rel(pre_test, post_test)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the pre-test and post-test scores.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the pre-test and post-test scores.")

plt.figure(figsize=(10, 6))

bar_width = 0.35
x = np.arange(len(pre_test))
plt.bar(x - bar_width/2, pre_test, bar_width, label='Pre-Test', color='blue')
plt.bar(x + bar_width/2, post_test, bar_width, label='Post-Test', color='orange')

plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Pre-test vs Post-test Scores Comparison')
plt.legend()

plt.tight_layout()
plt.show()

differences = post_test - pre_test
plt.figure(figsize=(10, 6))

plt.hist(differences, bins=10, edgecolor='black', alpha=0.7, color='green')

plt.axvline(np.mean(differences), color='red', linestyle='dashed', linewidth=2, label=f'Mean difference = {np.mean(differences):.2f}')

plt.xlabel('Differences (Post-test - Pre-test)')
plt.ylabel('Frequency')
plt.title('Distribution of Differences Between Pre-test and Post-test Scores')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):

    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    degrees_of_freedom = sample_size - 1

    t_score = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    margin_of_error = t_score * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

ci_lower, ci_upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def margin_of_error(sample_data, confidence_level=0.95):

    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)

    degrees_of_freedom = sample_size - 1

    if sample_size > 30:
        z_score = stats.norm.ppf((1 + confidence_level) / 2)
        margin_of_error = z_score * (sample_std / np.sqrt(sample_size))
    else:
        t_score = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)
        margin_of_error = t_score * (sample_std / np.sqrt(sample_size))

    return margin_of_error

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

margin_error = margin_of_error(sample_data, confidence_level=0.95)

print(f"Margin of Error at 95% confidence level: {margin_error:.4f}")
```

```python
# 13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

def calculate_expected_frequencies(observed_data):

    observed = np.array(observed_data)

    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)

    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total

    return expected
observed_data = np.array([[30, 10],
                          [20, 40]])

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
# 14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected):

    chi2_stat, p_value = chisquare(observed, expected)

    return chi2_stat, p_value


observed_data = np.array([12, 10, 9, 11, 8, 10])

expected_data = np.array([10, 10, 10, 10, 10, 10])

chi2_stat, p_value = goodness_of_fit_test(observed_data, expected_data)

print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data do not fit the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data fit the expected distribution.")
```

```python
# 16.  Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    n1 = len(sample1)
    n2 = len(sample2)
    if var1 > var2:
        f_statistic = var1 / var2
        df1 = n1 - 1
        df2 = n2 - 1
    else:
        f_statistic = var2 / var1
        df1 = n2 - 1
        df2 = n1 - 1

    p_value = 1 - f.cdf(f_statistic, df1, df2)
    print(f"F-statistic: {f_statistic}")
    print(f"P-value: {p_value}")

    if p_value < alpha:
        print("\nReject the null hypothesis: The variances are significantly different.")
    else:
        print("\nFail to reject the null hypothesis: The variances are not significantly different.")


sample1 = np.random.normal(loc=10, scale=2, size=30)
sample2 = np.random.normal(loc=10, scale=4, size=30)

f_test(sample1, sample2, alpha=0.05)
```

```python
# 17.  Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

import numpy as np
import scipy.stats as stats

def anova_test(*samples):

    f_statistic, p_value = stats.f_oneway(*samples)

    return f_statistic, p_value

group1 = np.random.normal(loc=10, scale=2, size=30)
group2 = np.random.normal(loc=12, scale=2, size=30)
group3 = np.random.normal(loc=14, scale=2, size=30)

f_statistic, p_value = anova_test(group1, group2, group3)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: At least one group mean is different.")
else:
    print("\nFail to reject the null hypothesis: The group means are not significantly different.")
```

```python
# 18.  Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def perform_anova(*groups):

    f_statistic, p_value = stats.f_oneway(*groups)

    return f_statistic, p_value

np.random.seed(42)
group1 = np.random.normal(loc=10, scale=2, size=30)
group2 = np.random.normal(loc=12, scale=2, size=30)
group3 = np.random.normal(loc=14, scale=2, size=30)

f_statistic, p_value = perform_anova(group1, group2, group3)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: At least one group mean is different.")
else:
    print("\nFail to reject the null hypothesis: The group means are not significantly different.")

plt.figure(figsize=(8, 6))
sns.boxplot(data=[group1, group2, group3], notch=True, patch_artist=True)
plt.xticks([0, 1, 2], ['Group 1', 'Group 2', 'Group 3'])
plt.title("Boxplot of Groups")
plt.ylabel("Values")
plt.show()
```

```python
# 19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_assumptions(*groups):

    assumptions = {}

    normality_results = []
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        normality_results.append((stat, p_value))

    assumptions['normality'] = normality_results

    stat, p_value = stats.levene(*groups)
    assumptions['equal_variance'] = (stat, p_value)

    return assumptions

def plot_qq(*groups):

    plt.figure(figsize=(12, 8))
    for i, group in enumerate(groups, start=1):
        plt.subplot(1, len(groups), i)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Q-Q Plot: Group {i}")
    plt.tight_layout()
    plt.show()


np.random.seed(42)
group1 = np.random.normal(loc=10, scale=2, size=30)
group2 = np.random.normal(loc=12, scale=2, size=30)
group3 = np.random.normal(loc=14, scale=2, size=30)

assumptions = check_assumptions(group1, group2, group3)

print("Normality Test Results (Shapiro-Wilk Test):")
for i, (stat, p_value) in enumerate(assumptions['normality'], start=1):
    print(f"Group {i} - Stat: {stat:.4f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print(f"  Reject null hypothesis: Group {i} is not normally distributed.")
    else:
        print(f"  Fail to reject null hypothesis: Group {i} is normally distributed.")

stat, p_value = assumptions['equal_variance']
print("\nLevene's Test for Equal Variance:")
print(f"Stat: {stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject null hypothesis: The variances are equal.")

plot_qq(group1, group2, group3)
```

```python
# 20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

factor_A = np.array(['Treatment 1', 'Treatment 2', 'Treatment 3'])
factor_B = np.array(['Time 1', 'Time 2', 'Time 3'])

data = []

for a in factor_A:
    for b in factor_B:
        performance = np.random.normal(loc=10 + (factor_A.tolist().index(a) * 2) + (factor_B.tolist().index(b) * 1), scale=2, size=20)
        for value in performance:
            data.append([a, b, value])

df = pd.DataFrame(data, columns=['Factor_A', 'Factor_B', 'Performance'])

formula = 'Performance ~ Factor_A * Factor_B'
model = smf.ols(formula, data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Factor_A', y='Performance', hue='Factor_B', data=df)
plt.title('Two-Way ANOVA: Interaction between Factor A and Factor B')
plt.show()
```

```python
# 21.  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def plot_f_distribution(d1, d2, title="F-distribution", xlabel="F-value", ylabel="Probability Density"):

    x = np.linspace(0, 5, 1000)

    y = stats.f.pdf(x, d1, d2)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'F-distribution (d1={d1}, d2={d2})', color='blue')
    plt.fill_between(x, 0, y, where=(x > stats.f.ppf(0.95, d1, d2)), color='red', alpha=0.3, label='Rejection Region')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

d1 = 3
d2 = 20
plot_f_distribution(d1, d2)
```

```python
# 24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

sample_size = 50
population_variance = 25
sample_data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)

sample_variance = np.var(sample_data, ddof=1)
print(f"Sample variance: {sample_variance}")
chi_square_statistic = (sample_size - 1) * sample_variance / population_variance

df = sample_size - 1

p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

alpha = 0.05
print(f"Chi-square statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from the hypothesized population variance.")

x = np.linspace(0, 50, 1000)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Chi-square Distribution (df=49)", color='blue')
plt.axvline(chi_square_statistic, color='red', linestyle='--', label=f"Chi-square Statistic = {chi_square_statistic:.2f}")
plt.fill_between(x, 0, y, where=(x > chi_square_statistic), color='red', alpha=0.3, label='Rejection Region')

plt.title("Chi-square Test for Population Variance")
plt.xlabel("Chi-square Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

successes_1 = 120
n_1 = 200
successes_2 = 150
n_2 = 200
p_hat_1 = successes_1 / n_1
p_hat_2 = successes_2 / n_2
p_pooled = (successes_1 + successes_2) / (n_1 + n_2)

se = np.sqrt(p_pooled * (1 - p_pooled) * (1 / n_1 + 1 / n_2))
z_statistic = (p_hat_1 - p_hat_2) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

alpha = 0.05
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Standard Normal Distribution", color='blue')
plt.axvline(z_statistic, color='red', linestyle='--', label=f"Z-statistic = {z_statistic:.2f}")
plt.fill_between(x, 0, y, where=(x > abs(z_statistic)), color='red', alpha=0.3, label='Rejection Region')

plt.title("Z-test for Proportions")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

group_1 = np.random.normal(loc=0, scale=np.sqrt(10), size=100)

group_2 = np.random.normal(loc=0, scale=np.sqrt(20), size=100)

var_1 = np.var(group_1, ddof=1)
var_2 = np.var(group_2, ddof=1)

F_statistic = var_1 / var_2

df1 = len(group_1) - 1
df2 = len(group_2) - 1

p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))


alpha = 0.05
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

x = np.linspace(0, 4, 1000)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="F-distribution (df1=99, df2=99)", color='blue')
plt.axvline(F_statistic, color='red', linestyle='--', label=f"F-statistic = {F_statistic:.2f}")
plt.fill_between(x, 0, y, where=(x > F_statistic), color='red', alpha=0.3, label='Rejection Region')

plt.title("F-test for Comparing Variances")
plt.xlabel("F-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

observed = np.random.randint(1, 7, size=1000)

observed_freq = np.array([np.sum(observed == i) for i in range(1, 7)])

expected_freq = np.full(6, 1000 / 6)

chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

alpha = 0.05
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies significantly differ from the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies do not significantly differ from the expected frequencies.")

categories = np.arange(1, 7)
plt.figure(figsize=(8, 6))

plt.bar(categories - 0.2, observed_freq, width=0.4, label='Observed', color='blue', align='center')
plt.bar(categories + 0.2, expected_freq, width=0.4, label='Expected', color='red', align='center')

plt.xlabel('Die Faces')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness of Fit: Observed vs Expected Frequencies')
plt.xticks(categories)
plt.legend()
plt.grid(True)
plt.show()
```
